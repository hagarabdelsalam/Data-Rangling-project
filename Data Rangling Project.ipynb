# Cleaning


```python
Supermarket_clean = Supermarket.copy()
```

```python
def determine_city(row):
    if row["Yangon"] == 1:
        return "Yangon"
    elif row["Naypyitaw"] == 1:
        return "Naypyitaw"
    else:
        return "Mandalay"

Supermarket_clean["city"] = Supermarket_clean.apply(determine_city, axis=1)
```

```python
Supermarket_clean
```

```python
Supermarket_clean = Supermarket_clean.drop(columns=["Yangon", "Naypyitaw", "Mandalay"])
```

```python
Supermarket_clean
```

```python
# Remove duplicate rows from the DataFrame
Supermarket_clean = Supermarket_clean.drop_duplicates()
```

```python
Supermarket_clean
```

```python
import numpy as np

# Update the 'Tax 5%' column: if the value is NaN or empty, calculate it; otherwise, keep the existing value.
Supermarket_clean['Tax 5%'] = np.where(Supermarket_clean['Tax 5%'].isna(), Supermarket_clean['Total'] * 0.05, Supermarket_clean['Tax 5%'])
```

```python
Supermarket_clean
```

```python
# test if there are any null value in the "Tax 5%" coulmn
Supermarket_clean[Supermarket_clean['Tax 5%'].isnull()]
```

```python
Supermarket_clean['Unit price'] = pd.to_numeric(Supermarket_clean['Unit price'], errors='coerce')
```

```python
print(Supermarket_clean.dtypes)
```

```python
# Update the 'Total' column: if the value is NaN or empty, calculate it; otherwise, keep the existing value.
Supermarket_clean['Total'] = np.where(Supermarket_clean['Total'].isna(), Supermarket_clean['Unit price'] * Supermarket_clean['Quantity']+Supermarket_clean['Tax 5%'] , Supermarket_clean['Total'])
```

```python
# test if there are any null value in the "Total" coulmn
Supermarket_clean[Supermarket_clean['Total'].isnull()]
```

```python
Supermarket_clean
```

```python
Supermarket_clean['Date'] = pd.to_datetime(Supermarket_clean['Date'], format='%m/%d/%Y')
```

```python
Supermarket_clean
```

```python
Supermarket_clean['Time'] = pd.to_datetime(Supermarket_clean['Time'], format='%H:%M', errors='coerce').dt.time
```

```python
Supermarket_clean
```

```python
Supermarket_clean[Supermarket_clean['Time'].isnull()]
```

```python
Supermarket_clean['Time'] = Supermarket_clean['Time'].fillna('8:30')
```

```python
Supermarket_clean
```

```python
Supermarket_clean['Time'] = Supermarket_clean['Time'].replace('8:30','8:30:00')
```

# Visualization 

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart for sales by city
city_sales = Supermarket_clean.groupby('city')['Total'].sum()
city_sales.plot(kind='bar', color=['#ff9999','#66b3ff','#99ff99'])
plt.title('Total Sales by City')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Bar chart for sales by product line
product_sales = Supermarket_clean.groupby('Product line')['Total'].sum()
product_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Convert Date to datetime if not already
Supermarket_clean['Date'] = pd.to_datetime(Supermarket_clean['Date'])

# Group sales by date
daily_sales = Supermarket_clean.groupby('Date')['Total'].sum()

# Line chart
plt.figure(figsize=(10, 6))
daily_sales.plot(kind='line', color='purple')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
# Pie chart for payment methods
payment_methods = Supermarket_clean['Payment'].value_counts()
payment_methods.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Payment Method Distribution')
plt.ylabel('')
plt.show()
```

```python
# Histogram for ratings
plt.figure(figsize=(8, 6))
sns.histplot(Supermarket_clean['Rating'], bins=10, kde=True, color='green')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot for sales by gender
plt.figure(figsize=(8, 6))
sns.boxplot(data=Supermarket_clean, x='Gender', y='Total', palette='pastel')
plt.title('Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales')
plt.show()
```
